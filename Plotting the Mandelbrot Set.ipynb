# Plotting the Mandelbrot Set

Write a Python script (in Jupyter Notebook) that plots the Mandelbrot set in the range [-2,1] on the real axis and [-1.5,1.5] on the imaginary axis.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the range for the plot
x_range = np.linspace(-2, 1, 1000)
y_range = np.linspace(-1.5, 1.5, 1000)
c = x_range[:, np.newaxis] + 1j * y_range[np.newaxis, :]
```

```python
# Initialize z and mask
z = np.zeros(c.shape, dtype=complex)
mask = np.zeros(c.shape, dtype=bool)
for i in range(50):  # Maximum iterations
   # Update z only where magnitude of z is less than 2
   mask = abs(z) <= 2
   z[mask] = z[mask] ** 2 + c[mask]
```

```python
# Convert final magnitudes into a boolean mask where True indicates inclusion in the set
final_mask = abs(z) <= 2
```

```python
# Plot the Mandelbrot set
plt.imshow(final_mask.T, extent=[-2, 1, -1.5, 1.5], origin='lower')
plt.gray()
plt.colorbar()  # Optional: Adds a color bar to indicate levels of divergence
plt.savefig('mandelbrot.png')
plt.show()
```
